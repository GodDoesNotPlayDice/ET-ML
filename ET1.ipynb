## Imports

```python
import pandas as pd
import seaborn as sbn
import numpy as np # linear algebra
import matplotlib.pyplot as plt
```

```python
# filename = 'Base_clientes_Monopoly.xlsx'
url = "https://github.com/razbackup/MonopolyBank/raw/main/Base_clientes_Monopoly.xlsx"

df = pd.read_excel(url)

# Muestra las primeras 10 filas
print(df.head(10))
```

```python
df.columns = df.iloc[0]

# Eliminar la primera fila para que no esté duplicada
df = df[1:].reset_index(drop=True)
df.head(10)
```

```python
df.columns.to_list()
```

```python
shape = df.shape
print(f"Cantidad de registros: {shape[0]}")
print(f"Cantidad de columnas {shape[1]}")
```

```python
columnas = df.columns

for col in columnas:
        try:
            df[col] = pd.to_numeric(df[col], errors='raise')
        except ValueError:
            print(col)
            pass
```

```python
for col in columnas:
  dtype = df[col].dtype
  columna = df[col]
  describe = columna.describe()
  print(f"Calculo estadistico basico para la columna {col}")
  print(describe)
  if dtype in ("int64","float64"):
    varianza = columna.var()
    print(f"Varianza: {varianza}")
  elif dtype == "object":
    unique = columna.unique()
    print(f"Valores Unicos: {unique}")
  print("\n\n")
```

```python
# Seleccionar solo las columnas numéricas
df_numerico = df.select_dtypes(include=['number'])

# Calcular la matriz de correlación
matriz_correlacion = df_numerico.corr()

# Mostrar la matriz de correlación
print(matriz_correlacion)
```
