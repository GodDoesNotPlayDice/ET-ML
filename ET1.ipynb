Como se puede observar obtenemos una tabla de 559 por 559 lo cual es ilegible para buscar correlaciones de manera manual, el plan a seguir es buscar entre todas las correlaciones ver cuales son mayores a 0.8 o menores a -0.8, para almacenarlas.
Para almacenarlas se creara un diccionario para las positivas y para las negativas pero seguiran la misma logica
La key del diccionario sera la columna que tiene la correlaciones, y el valor es una lista con todas las demas columnas que tiene una correlacion alta con esa columna.

```python
columnas_numericas = df.select_dtypes(include=['int64','float64'])
cors = columnas_numericas.corr()
high_correlation_pairs = []
correlaciones_dict_positivas = {}
correlaciones_dict_negativas = {}

# Iterar sobre la matriz de correlación
for i in range(cors.shape[0]):
    for j in range(i+1, cors.shape[1]):
        if abs(cors.iloc[i, j]) > 0.8:
            col1 = cors.columns[i]
            col2 = cors.columns[j]

            if cors.iloc[i, j] < 0:
                if col1 not in correlaciones_dict_negativas:
                    correlaciones_dict_negativas[col1] = []
                correlaciones_dict_negativas[col1].append((col2, cors.iloc[i, j]))

                if col2 not in correlaciones_dict_negativas:
                    correlaciones_dict_negativas[col2] = []
                correlaciones_dict_negativas[col2].append((col1, cors.iloc[i, j]))

            else:
                if col1 not in correlaciones_dict_positivas:
                    correlaciones_dict_positivas[col1] = []
                correlaciones_dict_positivas[col1].append((col2, cors.iloc[i, j]))

                if col2 not in correlaciones_dict_positivas:
                    correlaciones_dict_positivas[col2] = []
                correlaciones_dict_positivas[col2].append((col1, cors.iloc[i, j]))

```

Aca simplemente mostramos la cantidad de correlaciones positivas y negativas y con cuales se correlacionan

```python
count = 1
for columna, value in correlaciones_dict_negativas.items():
    print(count)
    print(f"Columna: {columna}, Correlaciones: {value}")
    count += 1
```

```python
count = 1
for columna, value in correlaciones_dict_positivas.items():
    print(count)
    print(f"Columna: {columna}, Correlaciones: {value}")
    count += 1
```

Al haber tantas correlaciones y no ser posible buscar de manera manual cual eliminar y cuales dejar, se penso que lo mejor es eliminar todas las columnas que tengan mas de 4 correlaciones altas, por que esto significaria que ya existen otras columnas que las representan. Y no de un numero menor, ya que desconocemos si eliminaremos columnas que se representen entre si por error, por lo cual no existan, dicho de otra forma desconocemos si eliminar 4 va a eliminar alguna relacion que no sabemos que existe.

```python
columnas_con_mas_cors = []

for columna, value in correlaciones_dict_positivas.items():
  if len(value) > 4:
    df.drop(columns=columna, inplace=True)
```

Estos son los imports necesarios para la identificación de los modelos

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.model_selection import RandomizedSearchCV
import warnings
from sklearn.exceptions import ConvergenceWarning
```

```python
varIndep = dfEscaled.drop(columns=["Renta"])
varDep = dfEscaled["Renta"]

X_train, X_test, y_train, y_test = train_test_split(varIndep, varDep, test_size=0.2, random_state=777)
print('Data set dividido')
```

El dataset se dividio en un 80% para ser utilizado para entrenar los modelos y un 20% para la fase de testing.
La variable a intentar predecir es la de renta de los clientes

```python

# Definir el rango de valores para los hiperparámetros
param_grid = {
    'alpha': np.arange(0.01, 3, 0.05),
    'max_iter': [i for i in range(1001,10001,500)]
}
print('Parametros generados')

# Configurar la búsqueda aleatoria
random_search = RandomizedSearchCV(Lasso(), param_grid, scoring='neg_mean_squared_error', cv=5, n_iter=75, random_state=777)
random_search.fit(X_train, y_train)


```

```python
print("Mejores parámetros:", random_search.best_params_)
print("Mejor puntaje MAE:", -random_search.best_score_)
```

Para poder generar el modelo de Regresión Lasso se utilizaron los siguientes parametros.
Para el alpha se genero una lista de entre 0.01 a 3(3 siendo excluyente), con un step de  0.05.
Para el max_iter que se refiere a las maximas iteraciones que va hacer el modelo para poder converger se genero una lista empezando desde 1001 hasta 10001 (10001 siendo excluyente) con un step de 500, por lo cual la lista queda del 1001 hasta el 9501 con una diferencia de 500 entre cada valor.

Para este modelo se eligio trabajar con 5 folders y solo 75 modelos de los posibles, no aumentamos en la cantidad de modelos o folders debido a que si tenemos un numero mayor el tiempo de ejecución sobrepasa la hora, lo cual no tenemos la disponibilidad para ello.

De entre los 75 modelos aleatoreos que se generados el mejor termino teniendo un alpha de 0.16 y un max_iter de 1501.

## Arboles de Regresion

```python
from sklearn.tree import DecisionTreeRegressor

warnings.simplefilter(action='ignore', category=ConvergenceWarning)
# Definir el rango de valores para los hiperparámetros del árbol de decisión
param_grid_tree = {
    'max_depth': [None] + list(range(5, 21, 5)),  # Sin límite o valores específicos
    'min_samples_split': range(2, 21, 2),  # Tamaño mínimo de muestras para dividir un nodo
    'min_samples_leaf': range(1, 21, 2),  # Tamaño mínimo de muestras en una hoja
    'max_features': ['auto', 'sqrt', 'log2', None]  # Selección de características
}

# Configurar la búsqueda aleatoria para el árbol de decisión
random_search_tree = RandomizedSearchCV(
    DecisionTreeRegressor(random_state=777),
    param_distributions=param_grid_tree,
    scoring='neg_mean_absolute_error',
    cv=5,
    n_iter=75,
    random_state=777
)

# Entrenar la búsqueda aleatoria
random_search_tree.fit(X_train, y_train)

# Imprimir los mejores parámetros y el mejor puntaje

```


Definicion de los parametros utilizados:


*   max_depth: es una lista que contiene numeros entre el 5 al 21 (21 siendo excluyente) con un step de 5, ademas de tener el valor de None, para que el arbol no tenga un max_depth.
*   min_samples_split: Es una lista de numeros entre el 2 al 21 (21 siendo excluyente) con un step de 2.


*   min_samples_leaf: Es una lista de numeros entre el 2 al 21 (21 siendo excluyente) con un step de 2.
*   Para max_features: Es una lista con 4 valores posibles.













Para el arbol de regresion tambien se generaron 75 modelos aleatoreos con 5 folders.

## SVM

```python
from sklearn.svm import SVC

# Definir el rango de valores para los hiperparámetros de SVM
param_grid_svm = {
    'C': np.logspace(-2, 10, 13),  # Penalización, valores en una escala logarítmica
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],  # Tipo de kernel
    'degree': range(2, 6),  # Grado del polinomio (para el kernel 'poly')
    'gamma': ['scale', 'auto'],  # Coeficiente del kernel (para 'rbf', 'poly' y 'sigmoid')
    'class_weight': ['balanced', None]  # Pesos de clase
}

# Configurar la búsqueda aleatoria para SVM
random_search_svm = RandomizedSearchCV(
    SVC(random_state=777),
    param_distributions=param_grid_svm,
    scoring='accuracy',
    cv=2,
    n_iter=5,
    random_state=777
)

# Entrenar la búsqueda aleatoria
random_search_svm.fit(X_trainCla, y_trainCla)
```

```python
# Imprimir los mejores parámetros y el mejor puntaje
print("Mejores parámetros para SVM:", random_search_svm.best_params_)
print("Mejor puntaje de precisión para SVM:", random_search_svm.best_score_)
```


Parametros:


*   C: Controla la penalización para los errores de clasificación en el modelo, los valores disponibles son entre 0.01 y 10^10 en una escala logarítmica.
*   Kernel: especifica la función de transformación de datos en el espacio de entrada para mejorar la separabilidad de las clases.
*   Degree: Controla el grado del polinomio cuando se utiliza el kernel poly. A mayor grado, más complejo el límite de decisión.
*   Gamma: Controla la influencia de cada punto de datos en el modelo en los kernels rbf, poly y sigmoid. Un valor alto de gamma hace que cada punto tenga una influencia más localizada, generando curvas de decisión más complejas, mientras que un valor bajo aumenta el radio de influencia de cada punto.
*   Class_weight: Ajusta el peso de cada clase, útil en problemas con clases desbalanceadas.




En este caso no pudimos ejecutar mas de 5 modelos con 2 folders, pero dentro de estos 5 modelos aleatoreos el mejor de ellos termino teniendo una Accuracy del 91% lo cual es algo bastante prometedor.

# Clustering Jerarquico

En este punto se va a probar algo muy similar a lo realizado en el Kmeans, la mayor diferencia es que solo se ejecuto con el MinMaxScaler, ya que con el StandardScaler el codigo no corrio en ningun momento.

```python
from itertools import product
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
import numpy as np

# Crear las combinaciones de affinity y linkage
affinities = ['euclidean', 'manhattan', 'cosine', 'precomputed']
linkages = ['ward', 'complete', 'average', 'single']

# Generar todas las combinaciones posibles
combinations = list(product(affinities, linkages))



print(combinations)
```

Se generaron las 16 posibles combinaciones de affinities y linkages para generar el modelo de todas estas, y ver cual es mejor utilizando el silhouette como metrica.

```python
varDepCla = dfNoEscalado["target"]

X_trainClJ, X_testClJ, y_trainClJ, y_testClJ = train_test_split(
    dfMinMax, varDepCla, test_size=0.2, random_state=777, stratify=varDepCla
)
print('Data set dividido')
```

El dataset se dividio entre training y testing en un 80% y 20% respectivamente, donde varDepCla es la variable a clasificar.

```python
# Lista para almacenar los resultados
results = []

# Iteramos sobre todas las combinaciones de affinity y linkage
for af, li in combinations:
    try:
        # Creamos el modelo de clustering
        model = AgglomerativeClustering(n_clusters= 2, metric=af, linkage=li)
        print('Modelo Creado')

        # Aplicamos el clustering
        labels = model.fit_predict(X_trainClJ)
        print("Modelo Entrenado")
        
        # Evaluamos el modelo utilizando Silhouette Score (métrica para medir la cohesión)
        if len(set(labels)) > 1:  # Solo calculamos la puntuación de la silueta si hay más de un clúster
            silhouette_avg = silhouette_score(X_trainClJ, labels)
            print("Silhouette Calculado")
        else:
            silhouette_avg = -1  # Si solo hay un clúster, la puntuación es -1 (no válida)
            print("Silhouette negativo")
        
        # Almacenamos la combinación y la puntuación
        results.append((af, li, silhouette_avg, labels))
        print("Modelo Agregado a resultados")
        
    except Exception as e:
        # En caso de error (por ejemplo, si affinity='precomputed' y no se pasó una matriz de distancias),
        # lo manejamos para que el código siga funcionando.
        print(f"Error con affinity={af} y linkage={li}: {e}")
```
