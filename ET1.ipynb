## Breve contexto del caso

El banco Monopoly, con presencia en Chile, ha sido adquirido recientemente por un banco extranjero llamado Dormammu. Como parte de la integración, Dormammu ha encargado a sus ingenieros realizar un análisis detallado del comportamiento de los clientes del banco Monopoly para detectar patrones y definir estrategias para abordar este nuevo mercado.

Para este propósito, los ingenieros de Monopoly han extraído una base de datos con información de clientes, que incluye 12 meses de datos mensuales por cliente. Esta base contiene 574 variables y 51,124 registros, con cada cliente representado una sola vez y las variables mensuales agregadas como columnas adicionales.

El objetivo del análisis es limpiar los datos, seleccionar variables relevantes y proporcionar insights a los nuevos dueños de Dormammu para entender mejor a los clientes y preparar estrategias de mercado efectivas.

```python
# Librerias
import pandas as pd
import seaborn as sns
import numpy as np # linear algebra
import matplotlib.pyplot as plt
```

```python
# Dos opciones de importar el archivo.

# filename = 'Base_clientes_Monopoly.xlsx'
url = "https://github.com/razbackup/MonopolyBank/raw/main/Base_clientes_Monopoly.xlsx"
df = pd.read_excel(url, header=1)
# Muestra las primeras 10 filas del dataset.
df.head(10)
```

```python
df.drop(['Unnamed: 574', 'Id', 'target'], axis=1, inplace=True)
```

Se dropean columnas irrelevantes para el estudio de los datos como lo son "target", "Unamed : 574" y "Id"

```python
shape = df.shape
print(f"Cantidad de registros: {shape[0]} clientes del banco.")
print(f"Cantidad de columnas {shape[1]}")
```

Las filas representan un cliente del banco con 12 meses de datos, y las columnas representan informacion del cliente y variables mensuales en su mayoría tratan sobre "comportamiento mensual en los productos financieros.

 Entonces podemos decir que el dataframe tiene 51,124 registros de clientes y 574 características.

```python
df_objetct = df.select_dtypes(include=['object'])
df_integers = df.select_dtypes(include=['int64'])
df_floats = df.select_dtypes(include=['float64'])
print(df_objetct.columns)
print(df_integers.columns)
print(df_floats.columns)
```

```python
columnas = df.columns

for col in columnas:
  dtype = df[col].dtype
  columna = df[col]
  describe = columna.describe()
  print(f"Calculo estadistico basico para la columna {col}")
  print(describe)
  if dtype in ("int64","float64"):
    varianza = columna.var()
    print(f"Varianza: {varianza}")
  elif dtype == "object":
    unique = columna.unique()
    print(f"Valores Unicos: {unique}")
  print("\n\n")
```

```python
columnas_numericas = df.select_dtypes(include=['int64']) # Aun asi pienso que deberiamos sacar las columnas que tienen descripciones
columnas_numericas.corr()
```

**Correlaciones altamente negativas o positivas:**

_Monoproducto y Ctacte_:
- **Correlación negativa (-0.91)**: Según los datos, los clientes que son **monoproducto** (solo poseen una tarjeta de crédito) tienen una **alta probabilidad de no tener una cuenta corriente** (Ctacte). Esto es consistente con la lógica, ya que los clientes que son monoproducto tienden a utilizar solo un tipo de producto financiero, en este caso una tarjeta de crédito, y no requieren de una cuenta corriente adicional.
  
_UsoL2 T01 y ColL2AC T01_:
- **Correlación positiva (0.95)**: Estas variables están muy relacionadas, lo que significa que miden aspectos muy similares. **UsoL2** se refiere al **uso de la línea de crédito para avances en cuotas** y **ColL2AC** probablemente mida colocaciones relacionadas con el mismo producto. Esto sugiere que ambas variables están midiendo el mismo comportamiento o producto financiero (avances en cuotas), lo que explica la alta correlación.

**Correlaciones Altas (Positivas):**

_FacDebCom T01 y TxsDebCom T01_:
- **Correlación positiva (0.84)**: Estas dos columnas miden las **facturaciones** y las **transacciones** relacionadas con compras mediante **tarjetas de débito comercial**. Dado que ambos indicadores están relacionados con el uso de la tarjeta de débito, es lógico que tengan una alta correlación. A medida que aumentan las transacciones, es natural que también aumenten las facturaciones.

_FacDebAtm T01 y TxsDebAtm T01_:
- **Correlación positiva (0.71)**: Estas variables también están relacionadas, midiendo las **facturaciones** y **transacciones** en **cajeros automáticos** mediante tarjeta de débito. Al igual que en el caso anterior, esta correlación alta es esperable, ya que un mayor número de transacciones en cajeros automáticos suele estar acompañado de un aumento en la facturación.

_FacCCPC T01 y TxsCCPC T01_:
- **Correlación moderadamente alta (0.53)**: Estas columnas se refieren a las **facturaciones** y **transacciones** realizadas en **compras en cuotas** a precio contado con tarjeta de crédito. Esta correlación indica que, aunque las dos variables están relacionadas, la facturación y las transacciones no siempre aumentan en la misma proporción, lo cual es comprensible dado que las compras a cuotas pueden tener comportamientos distintos (como montos más grandes o pagos diferidos).

**Correlaciones Moderadas:**

_Edad y Antigüedad_:
- **Correlación positiva (0.52)**: La edad y la antigüedad de los clientes están moderadamente correlacionadas, lo cual es esperado. A medida que los clientes envejecen, es más probable que hayan sido clientes del banco por más tiempo. Sin embargo, esta correlación no es muy alta, lo que indica que no todos los clientes mayores tienen alta antigüedad; es posible que algunos clientes más mayores se hayan unido al banco recientemente.

_Subsegmento y Edad_:
- **Correlación negativa (-0.52)**: La relación negativa entre el **subsegmento** (probablemente una clasificación de los clientes según su perfil o nivel de servicio) y la **edad** sugiere que los clientes más jóvenes tienden a pertenecer a subsegmentos más altos. Esto podría reflejar que el banco clasifica a los clientes jóvenes en perfiles que requieren más atención o tienen mayor capacidad de crecimiento.

**Correlaciones Bajas o No Significativas:**

- Muchas de las columnas muestran correlaciones muy bajas, lo que indica que no hay una relación lineal fuerte entre ellas. Esto es común cuando las variables no están directamente relacionadas o cuando miden diferentes aspectos del comportamiento financiero de los clientes. Por ejemplo, tener una cuenta de débito no necesariamente está correlacionado con tener un crédito hipotecario, lo que resulta en una correlación cercana a 0.


```python
# Grafico Desviacion estandar de la antiguedad de los clientes en el banco.
std_antiguedad = df['Antiguedad'].std()
media = df['Antiguedad'].mean()
mediana = df['Antiguedad'].median()

sns.histplot(df['Antiguedad'], kde=True)
plt.axvline(media, color='g', linestyle='-')
plt.axvline(mediana, color='b', linestyle='-')

plt.title('Histograma de la antiguedad de los clientes')
plt.xlabel('Antiguedad')
plt.ylabel('Frecuencia')
plt.show()


print(f"Media de la antiguedad de los clientes: {round(media)}")
print(f"Mediana de la antiguedad de los clientes: {round(mediana)}")
```

**Eje X (Antigüedad):** Representa la antigüedad de los clientes en meses, es decir, cuánto tiempo llevan siendo clientes del banco.

**Eje Y (Frecuencia):** Muestra la cantidad de clientes que tienen cada nivel de antigüedad. Las barras representan el número de clientes que caen dentro de cada rango de antigüedad.


La mayoría de los clientes tienen una antiguedad baja, concentrándose principalmente en los primeros 50 meses.

A medida que aumenta la antiguedad, la frecuencia de clientes disminuye significativamente, mostrando una tendencia decreciente y una cola larga hacia la derecha. Esto indica que hay muy pocos clientes con muchos años de antiguedad, se puede decir que muy pocos clientes que llevan mucho tiempo con el banco, lo cual podría indicar una alta tasa de renovación de la base de clientes o dificultades en la retención a largo plazo y por ultimo la alta frecuencia al inicio indica que el banco tiene una gran cantidad de clientes relativamente nuevos.

Este patrón podría sugerir la necesidad de estrategias de fidelización para retener clientes a lo largo del tiempo, ya que la mayoría de los clientes no alcanzan altas antiguedades.

```python
# Grafico de edad de los clientes en el banco
sns.histplot(df['Edad'], kde=True)
media = df['Edad'].mean()
mediana = df['Edad'].median()
plt.axvline(media, color='g', linestyle='-')
plt.axvline(mediana, color='b', linestyle='-')
plt.title('Histograma de la edad de los clientes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

print(f"Media de la edad de los clientes: {round(media)}")
print(f"Mediana de la edad de los clientes: {round(mediana)}")
print(f"Cantidad de clientes mayores a 36 años: {df[df['Edad'] > 36].shape[0]}")
print(f"Cantidad de clientes menores a 36 años: {df[df['Edad'] < 36].shape[0]}")
```

**Eje X (Edad):** Representa la edad de los clientes, desde los más jóvenes hasta los más mayores.

**Eje Y (Frecuencia):** Muestra la cantidad de clientes dentro de cada rango de edad.

La mayoría de los clientes se concentran entre los 20 y 40 años, con un pico notable alrededor de los 25-30 años. Esto sugiere que la base de clientes está compuesta principalmente por adultos jóvenes y adultos, pero no tanto de adultos mayores

A partir de los 40 años, la frecuencia de clientes disminuye gradualmente, mostrando una cola larga hacia la derecha, que se extiende hasta edades avanzadas, lo cual indica que hay algunos clientes de mayor edad, pero en menor cantidad.


```python
df['Antiguedad (Años)'] = df['Antiguedad'] / 12

# Gráfico de correlación con línea de tendencia
plt.figure(figsize=(10, 6))

# Scatterplot con transparencia para evitar sobrecarga de puntos
sns.scatterplot(data=df, x='Edad', y='Antiguedad (Años)', alpha=0.5)

# Línea de tendencia
sns.regplot(data=df, x='Edad', y='Antiguedad (Años)', scatter=False, color='red')

# Títulos y etiquetas
plt.title('Correlación entre Edad y Antiguedad (en años)')
plt.xlabel('Edad (Años)')
plt.ylabel('Antiguedad (Años)')

# Mostrar el gráfico
plt.show()
```

Si bien la correlacion no es muy alta sino moderada como se menciona arriba, la correlacion tampoco es perfecta en el sentido que hay adultos mayores con baja antiguedad en el banco.
