Como se puede observar, obtenemos una tabla de 559 por 559, lo cual es ilegible para buscar correlaciones de manera manual. El plan a seguir es revisar todas las correlaciones y identificar aquellas que sean mayores a 0.8 o menores a -0.8 para almacenarlas.

Para almacenar estas correlaciones, se creará un diccionario para las positivas y otro para las negativas, pero seguirán la misma lógica. La clave del diccionario será la columna que tiene las correlaciones, y el valor será una lista con todas las demás columnas que tengan una correlación alta con esa columna.

```python
columnas_numericas = df.select_dtypes(include=['int64','float64'])
cors = columnas_numericas.corr()
high_correlation_pairs = []
correlaciones_dict_positivas = {}
correlaciones_dict_negativas = {}

# Iterar sobre la matriz de correlación
for i in range(cors.shape[0]):
    for j in range(i+1, cors.shape[1]):
        if abs(cors.iloc[i, j]) > 0.8:  
            col1 = cors.columns[i]
            col2 = cors.columns[j]
            
            if cors.iloc[i, j] < 0:
                if col1 not in correlaciones_dict_negativas:
                    correlaciones_dict_negativas[col1] = []
                correlaciones_dict_negativas[col1].append((col2, cors.iloc[i, j]))

                if col2 not in correlaciones_dict_negativas:
                    correlaciones_dict_negativas[col2] = []
                correlaciones_dict_negativas[col2].append((col1, cors.iloc[i, j]))

            else:                     
                if col1 not in correlaciones_dict_positivas:
                    correlaciones_dict_positivas[col1] = []
                correlaciones_dict_positivas[col1].append((col2, cors.iloc[i, j]))

                if col2 not in correlaciones_dict_positivas:
                    correlaciones_dict_positivas[col2] = []
                correlaciones_dict_positivas[col2].append((col1, cors.iloc[i, j]))

```

Aquí simplemente mostramos la cantidad de correlaciones positivas y negativas, así como las columnas con las que se correlacionan.

```python
count = 1
for columna, value in correlaciones_dict_negativas.items():
    print(count)
    print(f"Columna: {columna}, Correlaciones: {value}")
    count += 1
```

```python
count = 1
for columna, value in correlaciones_dict_positivas.items():
    print(count)
    print(f"Columna: {columna}, Correlaciones: {value}")
    count += 1
```

Al haber tantas correlaciones y no ser posible determinar manualmente cuáles eliminar y cuáles conservar, se consideró que lo mejor es eliminar todas las columnas que tengan más de 4 correlaciones altas, ya que esto significaría que otras columnas ya las están representando.

```python
columnas_con_mas_cors = []

for columna, value in correlaciones_dict_positivas.items():
  if len(value) > 4:
    df.drop(columns=columna, inplace=True)
```

```python
# Lista de prefijos
prefijos = ['FlgActAN', 'FlgActCI', 'FlgActPAT', 'FlgActCCPC', 'FlgAct', 'FlgActCN',
            'FlgActAI', 'FlgActCCOT', 'FlgActCOL', 'Fac', 'Txs', 'FacCN', 'TxsCN',
            'FacAN', 'TxsCI', 'FacCI', 'TxsAN', 'FacAI', 'TxsAI', 'FacPAT', 'TxsPAT',
            'FacCCPC', 'TxsCCPC', 'FacCCOT', 'TxsCCOT', 'FacCOL', 'TxsCOL', 'FacDebCom',
            'TxsDebCom', 'FacDebAtm', 'TxsDebAtm', 'Col', 'ColL1T0', 'ColL1TE', 'ColL2T0',
            'ColL1TE', 'ColL2T0', 'ColL2AC', 'ColL2CC', 'ColMx', 'PagoNac', 'PagoInt',
            'EeccNac', 'EeccInt', 'UsoL1', 'UsoL2', 'UsoLI', 'IndRev', 'UsoL2', 'UsoLI']

# Rango de números
numeros = [str(i).zfill(2) for i in range(1, 13)]  # Genera ['01', '02', ..., '12']
# Iterar sobre los prefijos y números para reemplazar NaN por 0
for prefijo in prefijos:
    for numero in numeros:
        columna = f'{prefijo}_T{numero}'  # Construir el nombre de la columna
        if columna in df.columns:  # Verifica si la columna existe en el DataFrame
            df[columna] = df[columna].fillna(0)  # Reemplaza NaN por 0
            
```
